```julia
using DifferentialEquations
using LaTeXStrings
using Plots
```

```julia
f = (u, p, t) -> sin((t+u)^2)
u0 = -1.0
tspan = (0.0, 4.0)
```

```julia
ivp = ODEProblem(f, u0, tspan)
sol = solve(ivp, Tsit5());
```

```julia
plot(sol,label="solution",legend=:bottom,
    xlabel="t",ylabel=L"u(t)",title=L"u'=\sin((t+u)^2)")
```

```julia
@show sol(1.0);
```

```julia
[sol.t, sol.u]
```

```julia
# '!' mean that the parameter will be changed
scatter!(sol.t,sol.u,label="discrete values")
```

```julia
f = (u, p, t) -> (t+u)^2

ivp = ODEProblem(f, 1.0, (0., 1.))
sol = solve(ivp, Tsit5());
```

```julia
plot(sol, label="",
    xlabel=L"t",yaxis=(:log10,L"u(t)"), title="Finite-time blowup")
```

with $k=S=1$ and $ M = 0.25$, for the initial conditions $x(0) = 0.9M, 1.1M, 1.5M, 0.9S, 1.1S, 3S$. Show all the solutions together on one plot with $0 \leq x \leq 3$. 
(Note: One of the solutions will throw a warning and fail to reach , but you can plot it anyway.)

```julia
# Setting the parameter of the problem
k = S = 1;
M = 0.25;
# canditate_x0 = [0.9M, 1.1M, 1.5M, 0.9S, 1.1S, 3S];



# for value in canditate_x0
#     # Define f
#     f = (x,p,t) -> k*(S-x)*(x-M);
#     tspan = (0.0, 10.0);

#     # Setting the initial state
#     x0 = value;
    
#     # Define and solve the problem
#     ivp = ODEProblem(f, x0, tspan);
#     sol = solve(ivp, Tsit5());

#     # Plot the problem
#     plot!(sol)
# end

f = (x,p,t) -> k*(S-x)*(x-M);
tspan = (0.0, 10.0);

# Setting the initial state
x0 = 0.9M;

# Define and solve the problem
ivp = ODEProblem(f, x0, tspan);
sol = solve(ivp, Tsit5());

# Plot the solution
plot(sol, label="solution",legend=:bottom,
xlabel="t",ylabel=L"x(t)",title=L"x'=k(S-x)(x-M)")
```
