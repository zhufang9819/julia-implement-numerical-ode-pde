```julia
using Plots
using LaTeXStrings
using LinearAlgebra
using PrettyTables
using Printf
using DifferentialEquations
```

### 1. Problem Formulation
The FitzHugh–Nagumo equations are a simple model of the repeated firing of a neuron. They are given by
$$
\begin{split}\begin{split}
\frac{d v_1}{dt} &= - v_1(v_1-1)(v_1-a) - v_2 + I, \\
\frac{d v_2}{dt} &= \epsilon ( v_1 - \gamma v_2).
\end{split}\end{split}
$$
Assume $v_1(0) = 0,5$, $v_2(0) = 0.1$, $a = 0.1$, $\epsilon = 0.008$, $\gamma = 1$. For each value of $I$ below, find and plot the solution using 'solve' for $0 \leq t \leq 600$. The solutions are highly sensitive to $I$, and you need to change the requested absolute and relative error tolerances to $10^{-9}$. In each case the solution quickly approaches a periodic oscillation.

(a) $I=0.05527,\quad$ (b) $I=0.05683,\quad$ (c) $I=0.0568385,\quad$ (d)$ I=0.05740.$


### 2. Method
We use the 'solve' (in DifferentialEquations) to solve the problem.

#### 2.1 Implementation 

```julia
function fizhugh_nagumo(u, p, t)
    v₁, v₂ = u
    a, ϵ, γ, I = p
    return [-v₁*(v₁-1)*(v₁-a)-v₂+I, ϵ*(v₁-γ*v₂)]
end
```

```julia
# Setting parameters of the ivp
u₀ = [0.5, 0.1];
tspan = [0, 600];
a, ϵ, γ, I = 0.1, 0.008, 1, 0.05527;
ivp = ODEProblem(fizhugh_nagumo, u₀, tspan, [a, ϵ, γ, I])
sol = solve(ivp, Tsit5(), abstol=1e-9, reltol=1e-9);
```

```julia
plot(sol, vars=[1,2], label=[L"v_1" L"v_2"], title="FitzHugh–Nagumo model")
```

```julia
I = 0.05683;
ivp = ODEProblem(fizhugh_nagumo, u₀, tspan, [a, ϵ, γ, I])
sol = solve(ivp, Tsit5(), abstol=1e-9, reltol=1e-9);
plot(sol, vars=[1,2], label=[L"v_1" L"v_2"], title="FitzHugh–Nagumo model")
```
